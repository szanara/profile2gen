```python
import pandas as pd
import numpy as np
import os
from sklearn.model_selection import train_test_split
```

```python
dir2save = '/home/cz/mds3/data-centric/datasets/'
```

```python
name2save= os.path.join(dir2save, 'prepared/urinary')
```

```python
os.makedirs(name2save, exist_ok=True)
```

```python
def read_arf_file(arf_file_path):
    data = []
    with open(arf_file_path, 'r', encoding='utf-8') as file:
        for line in file:
            # Processar cada linha conforme necessário
            # Aqui, apenas adicionando cada linha aos dados
            data.append(line.strip())
    return data
```

```python
# Exemplo de uso
arf_data = read_arf_file(dir2save+"phpOAYun7.arff")
```

```python
data = arf_data[121:]
```

```python
attributes = arf_data[112:119]
```

```python
columns_extracted = []

# Analise cada string na lista de atributos
for attribute in attributes:
    # Separe o nome do atributo e o tipo de dados
    parts = attribute.split()
    attribute_name = parts[1]
    data_type = parts[2]
    columns_extracted.append(attribute_name)
```

```python
new_data = []
for i in range(len(data)):
    new_data.append(data[i].split(','))
```

```python
new_data = np.array(new_data)
```

```python
new_datafloat = []
for obtained in new_data:
    new_obtained = []
    for info in obtained:
        try:
            new_obtained.append(float(info))
        except ValueError:
            if info=='no':
                new_obtained.append(1)
            else:
                new_obtained.append(0)
    new_datafloat.append(new_obtained)
```

```python
new_datafloat = np.array(new_datafloat)
```

```python
new_datafloat.shape
```

```python
new_datafloat
```

```python
df= pd.DataFrame(new_datafloat, columns = columns_extracted)
```

```python
df = df.rename(columns={'V1': 'target'})
```

```python
df
```

```python
df_train, df_test = train_test_split(df, test_size=0.33, random_state=42)
```

```python
df_train.to_csv(os.path.join(name2save, 'urinary_train.csv'))
```
